```python
driver.get("https://www.naukri.com/")
```

```python
search_job=driver.find_element_by_id("qsb-keyword-sugg")
search_job.send_keys("Data Analyst")
search_location=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_location.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
# Searched for Data Analyst jobs in Bangalore
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
exp_salary_city=[i.get_attribute('title')for i in driver.find_elements_by_xpath('//span[@class="ellipsis fleft fs12 lh16"]')]
#This variable has stored experience, salary and city for all the 20 data instances in the page 
exp_salary_city=np.reshape(exp_salary_city,(20,3))
href_list=[i.get_attribute('href') for i in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')]
# This list of web elements has the links to jobs pages
job_titles=[]
job_descriptions=[]
exp_req=[]
salary=[]
city=[]
company_name=[]
for j in range(0,10):
    driver.get(href_list[j])
    # go to the jobs page
    try:
        job_titles.append(driver.find_element_by_xpath('//h1[@class="jd-header-title"]').get_attribute('title'))
        job_descriptions.append(driver.find_element_by_xpath('//div[@class="dang-inner-html"]').text.replace('\n',''))
        company_name.append(driver.find_element_by_xpath('//a[@class="pad-rt-8"]').text)
        # if job postings are normal webpage listing of the site then this will do just fine if not then the next part will execute
    except NoSuchElementException as e:
        job_titles.append(driver.find_element_by_xpath('//h1[@class="av-special-heading-tag " ]').text)
        job_descriptions.append(driver.find_element_by_xpath('//div[@class="clearboth description"]').text.replace('\n',''))
        company_name.append(driver.find_element_by_xpath('//p[@class="cpName f14"]').text)            
    exp_req.append(exp_salary_city[j,0])
    salary.append(exp_salary_city[j,1])
    city.append(exp_salary_city[j,2])

data_naukri_Data_Analyst_Bangalore={'Job_Title':job_titles,'Company_Name':company_name,'Experience_Required':exp_req,'Salary':salary,'Place':city,'Job_Description':job_descriptions}
data=pd.DataFrame(data_naukri_Data_Analyst_Bangalore)
data
#saved it in a dataframe
```

```python
driver.get("https://www.flipkart.com/")
```

```python

search_job=driver.find_element_by_xpath('//input[@title="Search for products, brands and more"]')
search_job.send_keys("sunglasses")
#search_location=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
#search_location.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
company_name=[]
product_description=[]
price=[]
discount=[]
m=0
while m<5:
    try:
        
        name_tags=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
        product_tags=driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
        price_tags=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
        discount_tags=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]/span')
        for i in name_tags:
            company_name.append(i.text)
        for j in product_tags:
            product_description.append(j.get_attribute('title'))
        for k in price_tags:
            price.append(k.text)
        for l in discount_tags:
            discount.append(l.text)
        next_button=driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
        next_button.click()
    except StaleElementReferenceException as e:
        WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"_3Ay6Sb")))
    m=m+1
```

```python
brand=company_name[0:100]
description=product_description[0:100]
Price=price[0:100]
Discount=discount[0:100]

data_flipkart_sunglass={'Brand':brand,'Description':description,'Price':Price,'Discount':Discount}
data=pd.DataFrame(data_flipkart_sunglass)
data
```

```python
page_list=[i.get_attribute('href') for i in driver.find_elements_by_xpath('//li[@class="pagination-active"]/a')]
next_pages=[i.get_attribute('href') for i in driver.find_elements_by_xpath('//li[@class="pagination-number"]/a')]
page_list=page_list+next_pages[0:2]
```

```python
rating=[]
review_sum=[]
review=[]
for page_tag in page_list:
    page_url=page_tag
    driver.get(page_url)
    rating_tags=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
    review_sum_tags=driver.find_elements_by_xpath('//h4[@class="product-product"]')
    review_tags=driver.find_elements_by_xpath('//span[@class="product-discountedPrice"]')
        #discount_tags=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]/span')
    for i in rating_tags:
        rating.append(i.text)
    for j in review_sum_tags:
        review_sum.append(j.text)
    for k in review_tags:
            review.append(k.text)        
```

```python
data_m_shoe={'Brand':rating[0:100],'Short_Shoe_Description':review_sum[0:100],'Price':review[0:100]}
data=pd.DataFrame(data_m_shoe)
data
```
